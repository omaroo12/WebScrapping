```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
```

```python
url='https://wuzzuf.net/search/jobs/?q=illustrator&a=navbl'
```

```python
client=urlopen(url)
```

```python
html=client.read()
```

```python
client.close()
```

```python
soub=bs(html,'html.parser')
soub
```

```python
containers=soub.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
```

```python
file=open('D:\insttant\dataAnalysisSheet\wuzzef.csv','w')
header='job_title,comp_name,job_type\n'
file.write(header)

```

```python
for container in containers:
    jobt=container.find_all('h2',{'class':'css-m604qf'})
    job_title=jobt[0].text.strip()
    
    compn=container.find_all('a',{'class':'css-17s97q8'})
    comp_name=compn[0].text.strip()
    
    jobtyp=container.find_all('div',{'class':'css-1lh32fc'})
    job_type=jobtyp[0].text.strip()
    
    file.write(job_title+","+comp_name+","+job_type+"\n")
    
```
